```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('data/m05/tracks_m05s01 in um per sec.csv')
df.columns=['Index', 'Track no', 'Slice no', 'X', 'Y', 'Distance', 'Velocity',
       'Pixel Value']
df.head()
```

```python
x_tracks=df.pivot_table(index='Track no', columns='Slice no', values='X')
print(x_tracks.shape)
x_tracks_long=x_tracks.dropna(subset=[1, 72])
print(x_tracks.shape)
print(x_tracks_long.shape)
```

```python
x_tracks_long.head()
```

```python
x_tracks_long.tail()
```

```python
y_tracks=df.pivot_table(index='Track no', columns='Slice no', values='Y')
y_tracks_long=y_tracks.dropna(subset=[1, 72])
y_tracks_long.head()
```

```python
x_tracks_long.to_csv('data/m05/x_tracks_m05s01 in um per sec.csv')
y_tracks_long.to_csv('data/m05/y_tracks_m05s01 in um per sec.csv')
```
